```python
df2 = pd.DataFrame({'Survived': df.Survived,
                   'Gender': df.Sex,
                    'Board': df.Embarked})
df2.tail()
```

```python
df5 = pd.concat([df2,df3], axis=0, sort=True) #row
df5.tail()
```

```python
df4.drop(['SibSp'], axis=1, inplace=True) #inplace save replace original
df4.tail()
```
